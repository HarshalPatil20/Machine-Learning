```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("Hiring.csv")
df.head(5)
```

```python
del df["Unnamed: 0"]
```

```python
df
```

```python
df.experience=df.experience.fillna(0)
```

```python
df
```

```python
import math
median_test_score = math.floor(df['test_score(out of 10)'].mean())
median_test_score
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)
df
```

```python
df.replace([{'five':5,'two':2,'seven':7,"three":3,"ten":10,"eleven":11}],inplace=True)
df
```

```python
x=df.drop(['salary($)'],axis='columns')
x
```

```python
y=df['salary($)']
y
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[7,7,6]])
```

```python
model.coef_
```
